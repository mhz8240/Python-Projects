```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('BSA_Application_Dataset_25.csv')
```

```python
df.iloc[:,:]
```

```python
df['scoring_efficiency_test'] = df['offensive_avg_points'] / (df['offensive_avg_field_goals_attempted'])
df['scoring_efficiency_test']

```

```python
df['threes_twos_ratio'] = df['offensive_avg_three_point_field_goals_attempted'] / df['offensive_avg_two_point_field_goals_attempted']
df['threes_twos_ratio']
```

```python
df['logo']
```

```python
df[['team','offensive_avg_three_point_field_goals_attempted','offensive_avg_two_point_field_goals_attempted']]
```

```python
from sklearn.metrics import r2_score

x = df['threes_twos_ratio']
y = df['scoring_efficiency_test']
m, b = np.polyfit(x, y, 1)
x_line = np.linspace(x.min()-0.001, x.max()+0.001, 200)
y_line = m * x_line + b

y_pred = x * m + b
r = r2_score(y, y_pred) ** 0.5

print("R value:", r)

```

```python
import matplotlib.image as mpimg
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
import urllib.request
from PIL import Image

fig, ax = plt.subplots(figsize=(10,7))
for i, row in df.iterrows():
  with urllib.request.urlopen(row['logo']) as url:
    img = Image.open(url)
    img = np.array(img)
  imagebox=OffsetImage(img, zoom=0.055)
  ab = AnnotationBbox(imagebox, (row['threes_twos_ratio'], row['scoring_efficiency_test']), frameon=False)
  ax.add_artist(ab)


ax.set_title('Offensive Scoring Efficiency vs 3PT-to-2PT Attempt Ratio', fontsize=16, weight='bold')
ax.set_xlabel('3PT-to-2PT Attempt Ratio', fontsize=14, weight='bold')
ax.set_ylabel('Scoring Efficiency', fontsize=14, weight='bold')
ax.grid(True, linestyle='--', alpha=0.6, which="major")
ax.set_ylim(1.00, 1.40)
ax.set_xlim(0.20, 0.70)
ax.plot(x_line, y_line, 'k--', label=f"r = {r:.2f}", alpha=0.75, c='tab:blue')
ax.legend(fontsize=12, edgecolor='black')
plt.tight_layout()
plt.show()
```
